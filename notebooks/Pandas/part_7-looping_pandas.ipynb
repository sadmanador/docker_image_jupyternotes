```python
import pandas as pd
```

# Looping over pandas data-frame

```python
brics = pd.read_csv('brics.csv', index_col = 0)
brics
```

```python
for label, row in brics.iterrows():
    print(label)
    print(row)
    print("..............")
```

```python
for label, row in brics.iterrows():
    print(label + " : "  + row['Capital'])
```

## Adding a new column in the table in efficient way

```python
brics["Name_length"] = brics["Country"].apply(len)
brics
```

```python
for label, row in brics.iterrows():
    brics.loc[label,'Characters_in_name'] = int(len(row['Country']))
brics
```
